# Data Pre-processing
- Data Type Correction
- Duplicate Handling
- Remove irrelevant columns
- Save cleaned data


```python
# Imports
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
# Load Data
DATA_PATH = r"C:\Users\hp\Downloads\BIA\Twitter_Sentiment_Analysis\Data\Twitter_Data.csv"
data= pd.read_csv(DATA_PATH)
data.head()
```

```python
# Data back-up
df = data.copy()
df.info()
```

```python
# Meta data
print("Number of rows: ",df.shape[0])
print("Number of columns:",df.shape[1])
all_columns= df.columns.to_list()
print("All columns: \n", all_columns)
```

```python
# Duplicates
print("Duplicate details: ", df.duplicated().sum())
print("Duplicated rows: \n")
df[df.duplicated()]
```

```python
# Check null values
df.isnull().sum()
```

```python
# Remove rows
df = df.dropna()
df.isnull().sum()
```

```python
# Drop irrelevant column
df.drop(columns = ["textID","selected_text"],axis= 1, inplace= True)
```

```python
# save cleaned data
CLEANED_DATA_PATH = r"C:\Users\hp\Downloads\BIA\Twitter_Sentiment_Analysis\Data\cleaned_data.csv"
df.to_csv(CLEANED_DATA_PATH, index= False)
```
