# Exploratory Data Analysis
- Statistical Analysis
- Visualization

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
CLEANED_DATA_PATH = r"C:\Users\hp\Downloads\BIA\Twitter_Sentiment_Analysis\Data\cleaned_data.csv"
data= pd.read_csv(CLEANED_DATA_PATH)
data.head()
```

```python
# Statistical Analysis
df= data.copy()
df.describe()
```

```python
# Checking Percentage distribution in Sentiment column
df['sentiment'].value_counts()
```

```python
# Target Distribution 
plt.figure(figsize=(6,4))

sns.countplot(x='sentiment', data=df)

plt.title("Sentiment Distribution", fontsize=14)
plt.xlabel("Sentiment", fontsize=12)
plt.ylabel("Count", fontsize=12)

plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```

```python
# Length of Tweets (Adding new columns)

# Number of characters
df['char_count'] = df['text'].apply(len)

# Number of words
df['word_count'] = df['text'].apply(lambda x: len(x.split()))

df.head()
```

```python
# Sentiment vs Word Count (Violin Plot)
plt.figure(figsize=(6,4))

sns.violinplot(x='sentiment', y='word_count', data=df)

plt.title("Sentiment vs Word Count")
plt.xlabel("Sentiment")
plt.ylabel("Word Count")

plt.show()
```

```python
# Sentiment vs Character (Violin Plot)
plt.figure(figsize=(6,4))

sns.violinplot(x='sentiment', y='char_count', data=df)

plt.title("Sentiment vs Character Count")
plt.xlabel("Sentiment")
plt.ylabel("Character Count")

plt.show()
```

```python
# Sentiment vs Word Count (Box Plot)
plt.figure(figsize=(6,4))

sns.boxplot(x='sentiment', y='word_count', data=df)

plt.title("Sentiment vs Word Count")
plt.xlabel("Sentiment")
plt.ylabel("Word Count")

plt.show()
```

```python
# Sentiment vs Character Count (Box Plot)
plt.figure(figsize=(6,4))

sns.boxplot(x='sentiment', y='char_count', data=df)

plt.title("Sentiment vs Character Count")
plt.xlabel("Sentiment")
plt.ylabel("Character Count")

plt.show()
```
